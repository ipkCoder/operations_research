```python
# retirement investing simulation
# assumes normal distribution for salary and investment return rates
import pandas as pd
import random
```

```python
age = 23
retire_age = 70
current_salary = 40000
current_portfolio = 0
contribution_rate = .1
employer_match = .04

raise_mean = .03
raise_stdev = .01
return_mean = .05
return_stdev = .12

trials = 300
```

```python
# number of rows
years = retire_age - age + 1

# dataframe for current trial
data = pd.DataFrame(np.zeros((years,7)),
                    columns=['BegPort','RaiseRate','Salary','InvAmnt','ReturnRate','PortEarn','EndPort'])

# dataframe for each trial
trial_end_data = pd.DataFrame(np.zeros((trials,6)),
                    columns=['AveRaiseRate','EndSalary','AveInvAmnt','AveReturnRate','AvePortEarn','EndPort'])
```

```python
for i in range(0,trials):
    
    for j in range(0,years):
        
        # set up begining profolio
        if j > 0:
            data.ix[j,'BegPort'] = data.ix[j-1,'EndPort']
        else:
            data.ix[j,'BegPort'] = current_portfolio
        
        # determine raise rate
        data.ix[j,'RaiseRate'] = random.normalvariate(raise_mean,raise_stdev)
        
        # increase salary by wage rate
        if j > 0:
            data.ix[j,'Salary'] = data.ix[j-1,'Salary']*(1 + data.ix[j,'RaiseRate'])
        else:
            data.ix[j,'Salary'] = current_salary
            
        # determine contribution amount
        data.ix[j,'InvAmnt'] = data.ix[j,'Salary']*(contribution_rate + employer_match)
        
        # determine return rate, total dollar return, and ending portfolio balance
        data.ix[j,'ReturnRate'] = random.normalvariate(return_mean,return_stdev)
        data.ix[j,'PortEarn'] = (data.ix[j,'BegPort'] + .5*data.ix[j,'InvAmnt'])*data.ix[j,'ReturnRate']
        data.ix[j,'EndPort'] = data.ix[j,'BegPort'] + data.ix[j,'InvAmnt'] + data.ix[j,'PortEarn']
    
    
    # update trial information
    trial_end_data.ix[i,'EndSalary'] = data.ix[years-1,'Salary']
    trial_end_data.ix[i,'AveRaiseRate'] = mean(data.ix[:,'RaiseRate'])
    trial_end_data.ix[i,'AveInvAmnt'] = mean(data.ix[:,'InvAmnt'])
    trial_end_data.ix[i,'AveReturnRate'] = mean(data.ix[:,'ReturnRate'])
    trial_end_data.ix[i,'AvePortEarn'] = mean(data.ix[:,'PortEarn'])
    trial_end_data.ix[i,'EndPort'] = data.ix[years-1,'EndPort']
    
```
